```python
from sunbird.data.data_readers import Patchy
import numpy as np
import matplotlib.pyplot as plt

s = np.load('/global/homes/e/epaillas/code/sunbird/data/s.npy')
smin, smax = 0.7, 150
s = s[(s > smin) & (s < smax)]
ells = [0]
ds = [0]

slice_filters = {
    's': [smin, smax],
}
select_filters = {
    'quintiles': ds,
    'multipoles': ells,
}

datavector = Patchy(
    statistics=['density_split_cross'],
    select_filters=select_filters,
    slice_filters=slice_filters,
).get_observation(phase=1)

fig, ax = plt.subplots()
ax.plot(s, s**2*datavector)

plt.show()
```
